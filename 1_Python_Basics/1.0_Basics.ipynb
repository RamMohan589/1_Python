### Indentation  

It is used to define the structure and hierarchy of the code. Unlike many other programming languages that uses braces {} to delimit the block of the code , python uses Indentaion to determine the grouping the statments. This means that all the statements with in a block must be intended to at the same level


```python
## Indentation -> Python uses indentation to define the block of the code. Consistent use of spaces(commonly 4) or a tab is required
age=32
if age>30:
    print(age)

print(age)

```

```python
## Line Continuation
## use back slash (\) to continue a statement to the next line
total = 1+2+3+4+5+6+7+ \
4+5+6

print(total)  ## here you will be able to see total from 1to7 along with 4+5+6 -> i.e if you give space and type back slash it will be line Continuation
 
```
